```python
import pandas as pd
```

```python
sample_df = pd.DataFrame({"name" : ["A", "B","C","D","E","F"],
                         "sex" : [0,0,1,0,1,1], 
                         "age" : [13, 35, 67, 46, 22, 20],
                         "hometown" : ["Japan", "America", "Japan", "China", "China", "America"]
                         })
sample_df
```

```python
sample_df.isnull()#今回は欠損値がないので全てFalseとなります
```

```python
sample_df2 = pd.DataFrame({"name" : ["G", "H","I","J","K","L"],
                         "sex" : [1,1,0,0,1,0], 
                         "age" : [15, 40, 78, 45, 19, 21],
                         "hometown" : ["America", "Japan", "Japan", "Canada", "China", "China"]
                         })
sample_df2
```

```python
pd.concat([sample_df, sample_df2])#シンプルに結合
```

```python
sample_df3 = pd.concat([sample_df, sample_df2], ignore_index=True)#ignore_index=Trueにしておくことでインデックスが再度振り分けられる。
sample_df3
```

```python
sample_df4 = pd.DataFrame({"job" : ["student", "doctor", "doctor", "engineer", "student", "student", "student", "researcher", "researcher", "policemen", "student", "student"]})
sample_df4
```

```python
sample_df5 = pd.concat([sample_df3, sample_df4], axis=1)
sample_df5 #axis=1にすることで横に結合することができる。
```

```python
filter_Japan = sample_df5["hometown"] == "Japan"#日本人だけのデータを抽出
sample_df5[filter_Japan]
```

```python
filter_Japan_researcher = (sample_df5["hometown"] == "Japan") & (sample_df5["job"] == "researcher")#日本人かつ研究者だけのデータを抽出
sample_df5[filter_Japan_researcher]
```

```python
filter_student_above_20 = (sample_df5["job"] == "student") & (sample_df5["age"] >= 20 )#学生で２０歳以上のデータだけを抽出
sample_df5[filter_student_above_20]
```

```python
sample_df5.sort_values(by=["age"], ascending=True)#年齢が低い順にソート。Falseにすると年齢が高い順にソートすることになる。
```

```python
def return_png(*args):
    png_list = []
    for filename in args:
        if filename[-4:] == ".png":
            png_list.append(filename)
        
    return png_list
```
