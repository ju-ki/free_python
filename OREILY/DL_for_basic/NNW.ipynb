```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
def step_function(x):
    y = x > 0
    return y.astype(np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-.1, 1.1)
plt.title("Step function")
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.title("Sigmoid function")
plt.show()
```
