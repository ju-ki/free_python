## 一般化線形モデル
* 正規分布以外の確率分布を使うことができる

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
from scipy import stats
sns.set()
%matplotlib inline
```

```python
stats.binom.pmf(k = 1, n = 2, p = 0.5)
```

```python
np.random.seed(1)
stats.binom.rvs(n = 10, p = 0.2, size = 5)
```

```python
binomial = stats.binom(n = 10, p = 0.2)
np.random.seed(1)
rvs_binomial = binomial.rvs(size = 10000)
m = np.arange(0, 10, 1)
pmf_binomial = binomial.pmf(k = m)
sns.distplot(rvs_binomial, bins = m, kde = False, norm_hist = True,
            color = 'gray')
plt.plot(m, pmf_binomial, color= 'black')
```

#### ポアソン分布
* カウントデータが従う離散型の確率分布
* 母数は強度のλのみ

```python
stats.poisson.pmf(k = 2, mu = 5)
```

```python
np.random.seed(1)
stats.poisson.rvs(mu = 2, size = 5)
```

```python
poisson = stats.poisson(mu = 2)
np.random.seed(1)
rvs_poisson = poisson.rvs(size = 10000)
pmf_poisson = poisson.pmf(k = m)
sns.distplot(rvs_poisson, bins = m, kde = False, norm_hist = True, color = 'gray')
plt.plot(m, pmf_poisson, color = 'black')
```

```python
N = 100000000
p = 0.00000002
binomial_2 = stats.binom(n = N, p = p)
pmf_binomial_2 = binomial_2.pmf(k = m)
plt.plot(m, pmf_poisson, color = 'gray')
plt.plot(m, pmf_binomial_2, color = 'black', linestyle = "dotted")
```
