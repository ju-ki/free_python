# 統計学の目的
1. 手持ちのデータを整理・要約すること（**記述統計**）
2. 手持ちのデータを用いて未知のデータの推測する（**推測統計**）

## 用語説明
**標本**

・手持ちデータのこと

**母集団**

・標本を含む全てのデータのこと


## 標本が得られるプロセス
**確率変数**

確率的な法則従って変化する値のこと。
確率変数における具体的な値のことを**実現値**と呼ぶ。
実際に得る事ができた値のこと。

ex:例えば釣りをした時10センチの魚を釣ったとします。その際の10センチというデータを実現値と呼ぶ。

**サンプリング**

母集団から標本を得ること。

**単純ランダムサンプリング**

母集団1つ1つの要素が無作為に選ばれる選び方。日本人の中から1人選ぶ行為がこれにあたる。


**サンプルサイズ**

標本の大きさのこと。

**全数調査と標本調査**

・ 母集団全体を調べることを**全数調査**

・ 母集団の一部のみを調べることを**標本調査**

P(170<=cm<172)=1/5

P(174<=cm<176)=1/5

P(176<=cm<178)=1/5

この状態の時、確率分布に従うと呼ぶ。

**無限母集団**

無限の大きさを持つ母集団のことを指す。

ex: サイコロ投げの結果（投げれ続ければ無限に母集団は増加していく）

### 母集団分布の推定

母集団分布を得る際、確率を即座に計算できる**確率質量変数**や**確率密度変数**

**正規分布**

計算が簡単で、データとよく対応する確率分布のこと

1. -∞〜∞の実数値をとる
2. 平均付近の確率密度が大きい
3. 平均から離れるほど、確率密度が小さくなる。
4. 確率密度は平均を中心として左右対称



### 確率質量関数と確率密度関数

**確率質量関数**

データ引数にとると、確率が計算結果として出てくる関数のこと。

**確率密度**

連続型の変数である場合計算ができない。数値の代わりに確率密度を使う。

ex:

身長が170cmになる確率を求める事ができないが、身長が１７０cm〜１７２cmになる確率であれば、確率密度を１７０〜１７２の間で積分する事ができる。

ここでの積分は**足し算の延長線**にあるもの。

１７０〜１７２cmは無限に存在する。ここでは足し算を使う事ができないので、積分を用いる。

**母数**

確率分布を特徴付ける値のこと。確率分布のパラメタ。

**母集団分布の推定＝分布の決定＋母数の推定**

1. 分布の形を決める
2. 母数の決定

母集団分布を正規分布と仮定した場合、次に母数の推定をしなければなりません。

母数を推定する方法として、標本の統計量を母集団分布の母数と考える事です。


### 統計量の計算

**標本**

$$\{x_i\}_{i=1}^N = \{x_1, x_2, ..., x_N\}$$

**平均値**

$$\mu = \frac{1}{N}\sum_{i=1}^N x_i$$

**期待値**

$$\mu = \sum_{i=1}^N P(x_i)*x_i$$

**分散（標本分散）**

$$\sigma^2 = \frac{1}{N}\sum_{i=1}^N (x_i-\mu)^2$$

$$(x_i-\mu)$$を**偏差**と呼ぶ

これの二乗が**偏差平方和**

また確率を用いて表すと

$$\sigma^2 = \frac{1}{N}\sum_{i=1}^N P(x_i)*(x_i-\mu)^2$$

$$(x_i-\mu)^2$$は期待値とデータの距離とみなす事ができる．

**不偏分散**

標本分散だと母分散と比べて分散を過小評価してしまうバイアスが生じる．
これを解消したものが不偏分散．

$$\sigma^2 = \frac{1}{N-1}\sum_{i=1}^N (x_i-\mu)^2$$


**標本分散が偏りを持つ理由**

ex:

湖に魚が７匹しかいなかったことにします．
{1,2,3,4,5,6,7}

これの母平均は４

サンプリングとして３匹つったとします．

{1,2,3}

これの標本平均は２

分散は平均とデータとの距離なので，本来は母平均からの距離を計算するべき．

なので不偏分散は標本分散より少し大きな値になるように補正されている．

**標準偏差**

$$\sigma = \sqrt\sigma^2 = \sqrt{\frac{1}{N-1}\sum_{i=1}^N(x_i-\mu)^2}$$

**絶対値を使わない理由**

絶対値を使うと計算が複雑になる．分岐が必要になったり，四則演算だけでは求める事ができない．
また微分も使う事ができないので，最大値や最小値を求めることができない．

### 確率論の基本

**集合の二つの表記方法**

｜や；の右側は条件を表す．

**標本点・標本空間・事象**

起こり得る可能な結果(**標本点**)

表し方：ω

標本点の全体集合(**標本空間**)

表し方：Ω

**確率の公理主義的定義**

(a)全ての事象Aに対してAは0から１

(b)標本空間＝１

(c)排反事象で確率の和で計算ができるいう事
