```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.plot([1,2,3,4])
plt.ylabel('Some number')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9 ,16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9 ,16], 'ro')
plt.axis([0, 6, 0, 10])
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
area = np.random.rand(50) *100
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
objects =('a', 'b', 'c', 'd', 'e', 'f')
y_pos = np.arange(len(objects))
values = [1, 2, 3, 4, 5, 6]

plt.bar(y_pos, values, alpha=0.5)
plt.xticks(y_pos, objects)
plt.title('Title')
plt.ylabel('Usage')
plt.show()
```

```python
labels = ('python', 'C#', 'Go', 'Java')
sizes = [10, 20, 30, 40]
colors = ['red', 'blue', 'green', 'yellow']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

```python
y = [2,4,6,8,10]
y2 = [10,11,12,13,14]
x = np.arange(5)
plt.plot(x, y, label='Y')
plt.plot(x, y2, label='Y2')
plt.legend()
plt.show()
```
