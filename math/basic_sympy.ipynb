```python
def calc_diff(equations):#関数化してみた
    x = Symbol('x')
    differentiated = diff(equations)
    return differentiated
```

```python
calc_diff(3*sin(x) - 4*cos(x))
```

```python
calc_diff(log(x) + log(2*x**2 + 1))
```

```python
calc_diff(log(x) / x**2)
```

```python
calc_diff(x**(-3*x))
```
