```python
def calc_diff(equations):#関数化してみた
    x = Symbol('x')
    differentiated = diff(equations)
    return differentiated
```

```python
def cacl_integarete(equation):
    x, C = symbols('x C')
    integrated = integrate(equation)
    return integrated
```

```python
cacl_integarete(1 / x)
```

```python
cacl_integarete(cot(x))
```
